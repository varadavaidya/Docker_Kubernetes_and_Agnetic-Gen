```python
import chromadb
```

```python
client= chromadb.Client()
```

```python
collection = client.get_or_create_collection(name= "my_collection")
```

```python
collection.add(
    ids=["id1","id2"],
    documents = [
        "this is a document about pineapples",
        "this is a document about orange"
    ]
)
```

```python
results = collection.query(
    query_texts= ["this is query about hawaii"],
    n_results = 2
)

print(results)
```

```python
results = collection.query(
    query_texts= ["this is query about pulp"],
    n_results = 2
)

print(results)
```

###  making Chroma DB persistent

```python
client_new = chromadb.PersistentClient(path = "./chroma")
```

```python
collection_new = client_new.get_or_create_collection(name = "persistent_db")
```

```python
collection_new.add(
    ids = ["id1"],
    documents = ["There out in the darkness, a fugitive running, fallen from god!"]
)
```

```python
collection_new.get(ids=["id1"],
                  include = ["embeddings"])
```
